# STEP #1: PROBLEM STATEMENT


- Predicting if the cancer diagnosis is benign or malignant based on several observations/features 
- 30 features are used, examples:
        - radius (mean of distances from center to points on the perimeter)
        - texture (standard deviation of gray-scale values)
        - perimeter
        - area
        - smoothness (local variation in radius lengths)
        - compactness (perimeter^2 / area - 1.0)
        - concavity (severity of concave portions of the contour)
        - concave points (number of concave portions of the contour)
        - symmetry 
        - fractal dimension ("coastline approximation" - 1)

- Datasets are linearly separable using all 30 input features
- Number of Instances: 569
- Class Distribution: 212 Malignant, 357 Benign
- Target class:
         - Malignant
         - Benign


https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic)

# STEP #4: MODEL TRAINING (FINDING A PROBLEM SOLUTION)

```python
# Let's drop the target label coloumns
X = df_cancer.drop(['target'],axis=1)
X
```

```python
y = df_cancer['target']
y
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
# tts Shift + Tab
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.20, random_state=5)
```

#### Q: It's a Supervised Machine Learning Problem. Why?
### A: We have labeled data which means some of the data is already tagged with the correct answer. Now the new dataset can be supervise or teach by the expericence of correct labled data which already takes place.



#### Q. Which technique can be used here for this problem? 
### A. This problem statement falls into Classification; means to group the output inside a class. Here, the algorithm tries to label input into two distinct classes, which are Malignant and Benign (The binary Classification).

##### Q. Which Classification would be better to use? 
### A. Support vector machine (SVM) is best suited for multimedia information retrieval, bioinformatics, pattern recognition, image classification, genes classsification, drug disambiguation, etc because of Non-linear classification with high-dimensional feature where sparsity is also high. 
### Mostly learning algorithm tries to learn the most common characteristics where as SVM works in the other way around, it finds the most similar examples between classes. Those will be the support vectors. This is best for Kernel trick and optimal margin based classification technique

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
svc_model.fit(X_train, y_train)
```

# STEP #5: EVALUATING THE MODEL

```python
y_pred = svc_model.predict(X_test)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot = True)
```

# STEP #6: IMPROVING THE MODEL 

### Preprocessing Data using MinMax Normalization Data
#### This estimator scales and translates each feature individually such that it is in the given range on the training set, e.g. between zero and one.

```python
min_train = X_train.min()
min_train
```

```python
range_train = (X_train - min_train).max()
range_train
```

```python
X_train_scaled = (X_train - min_train)/range_train
X_train_scaled
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue = y_train)
```

```python
sns.scatterplot(x = X_train_scaled['mean area'], y = X_train_scaled['mean smoothness'], hue = y_train)
```

```python
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test)/range_test
```

```python
svc_model_s = SVC()
svc_model_s.fit(X_train_scaled, y_train)
```

```python
y_pred_s = svc_model_s.predict(X_test_scaled)
cm_s = confusion_matrix(y_test, y_pred_s)
```

```python
sns.heatmap(cm_s, annot = True)
```

```python
print(classification_report(y_test, y_pred_s))
```

# IMPROVING THE MODEL - PART 2 
### Hyperparameter Optimization using Grid Search

```python
from scipy.stats import expon
param_grid = {'C':[0.1, 1,10,100], 'gamma':[0.1, 0.01, 0.001, 0.0001], 'kernel':['rbf'], 'class_weight': ['none','balanced']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_pred = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, grid_pred)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,grid_pred))
```
